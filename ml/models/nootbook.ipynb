**Visualiser l'impact des variable categorial sur la cible Attrition**

```python
from numpy import int64
import matplotlib.pyplot as plt

categorial_data=data[['BusinessTravel', 'DailyRate', 'Department', 'Education', 'EducationField', 'EmployeeCount', 'Over18','EnvironmentSatisfaction', 'Gender', 'JobInvolvement','JobLevel', 'JobRole', 'JobSatisfaction','MaritalStatus', 'RelationshipSatisfaction' ,'PerformanceRating','StockOptionLevel', 'WorkLifeBalance']]

data['Attrition']= data['Attrition'].map({'No':0, 'Yes':1})
```

```python
nemeric_data= data[['Attrition','DistanceFromHome', 'EmployeeNumber', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate', 'DailyRate','NumCompaniesWorked', 'OverTime', 'PercentSalaryHike', 'TotalWorkingYears', 'TrainingTimesLastYear','YearsAtCompany', 'YearsInCurrentRole','YearsSinceLastPromotion', 'YearsWithCurrManager']]
corr_matrix= nemeric_data.corr(numeric_only=True)
plt.figure(figsize=(20,20))
sns.heatmap(corr_matrix,annot=True)
plt.show()
```

**Boxplot : pour voire si les column numerique contient des outliers**

```python
#check if there is some outliers
nemeric_data_no_target= nemeric_data.drop(columns=['Attrition'])
plt.figure(figsize=(20,30))
for i, num_feature in enumerate(nemeric_data_no_target.columns):
    plt.subplot(6,3,i+1)
    sns.boxplot(y= num_feature, data= nemeric_data_no_target)
    plt.title(f'boxplot for {num_feature}')

plt.show()
```

**Visualiser l'impact des variable categorial sur la cible Attrition**

 - Jutilise le countplot pour les variables catgorielles parce que la variable cible (Attrition) est galement catgorielle.
 - Le countplot me permet de comparer la distribution des employs ayant quitt lentreprise et de ceux qui sont rests selon les diffrentes catgories de chaque variable.
 - En observant les diffrences de comptage entre les catgories, je peux identifier quelles variables catgorielles sont le plus fortement associes  lattrition.

```python
categorial_data = data[['JobRole', 'BusinessTravel', 'Department', 'Education', 'EducationField', 'EmployeeCount', 'EnvironmentSatisfaction', 'Gender', 'JobInvolvement','JobLevel', 'JobSatisfaction','MaritalStatus', 'RelationshipSatisfaction' ,'PerformanceRating' , 'OverTime','StockOptionLevel', 'WorkLifeBalance']]
for cat in categorial_data:
    plt.figure(figsize= (25,3))
    sns.countplot(x=data[cat], hue=data['Attrition'])
    plt.title(cat)
    plt.show()
```
